```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('Salary_dataset.csv')
df
```

```python
X=df[['YearsExperience']]
y=df[['Salary']]
```

```python
plt.scatter(X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Coefficient (Slope):", model.coef_)
print("Intercept:", model.intercept_)
print("MSE:", mse)
print("R² Score:", r2)
```

```python
#here it shows that
#1] Coefficent - For each extra year of experience, the model predicts the salary increases by about ₹9,413.
#2] Intercept - If someone has 0 years of experience, the model still predicts they’d earn about ₹25,417.96
#3] MSE - This is the average squared error between actual and predicted salaries.
#4] The model explains 47% of the variation in salary ,that means experience alone is not enough to fully explain salary differences.
```

```python
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, model.predict(X), color="red", label="Regression Line")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
